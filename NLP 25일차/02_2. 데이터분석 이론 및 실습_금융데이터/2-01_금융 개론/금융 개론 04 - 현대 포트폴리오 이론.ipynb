# μƒκ΄€κ³„μ correlation coefficient

* κ³µλ¶„μ‚°μ„ ν‘μ¤€νΈμ°¨ κ³±μΌλ΅ λ‚λ„μ–΄ ν‘μ¤€ν™”ν• κ°’
* μƒκ΄€κ³„μ(Ο: rhoλ΅ μ½μ)

$$ \rho_{AB} = \frac { \sigma_{AB} }{ \sigma_{A} \sigma_{B}}  $$

μƒκ΄€κ³„μμ‹μ„ μ μ©ν•λ©΄ ν¬νΈν΄λ¦¬μ¤ λ¶„μ‚°μ„ μ•„λμ™€ κ°™μ΄ μ •μν•  μ μλ‹¤

# ν¬νΈν΄λ¦¬μ¤μ μ„ν— κ³„μ‚°
ν¬νΈν΄λ¦¬μ¤μ λ¶„μ‚°(μ„ν—)μ€ κµ¬μ„±λ κ° μμ‚°μ λΉ„μ¨κ³Ό λ‹¤λ¥Έ μμ‚°μ λΉ„μ¨ κ·Έλ¦¬κ³  κ³µλ¶„μ‚°μ κ³±μΌλ΅ ν‘ν„

$$ \sigma^2_p = Var(r_p) = Var (w_Ar_A + w_Br_B) = w_A^2 \times \sigma_A^2 + w_B^2 \times \sigma_B^2 + 2 w_Aw_B\sigma_{AB}  $$

$$ \sigma_{π΄πµ} = π¶π‘π‘£(π‘_π΄, π‘_B) = E[{_π‘π΄ β’ πΈ(π‘_π΄)}{π‘_πµ β’ πΈ(π‘_πµ)}] \\

$$ \rho_{AB} = \frac { \sigma_{AB} }{ \sigma_{A} \sigma_{B}} = \frac { -508 }{ (23.69) (21.54) } = -0.996 $$

$$ \sigma_p^2 = 0.5^2 Γ— 561 + 0.5^2 Γ— 464 + 2 Γ— 0.5 Γ— 0.5 Γ— (β’508) = 2.25 $$
