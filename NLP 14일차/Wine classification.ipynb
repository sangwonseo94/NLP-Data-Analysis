```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as function
import torch.optim as optim
from torch.utils.data import DataLoader,TensorDataset

from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split

import pandas as pd
```

```python
wine =load_wine()
```

```python
pd.DataFrame(wine.data , columns=wine.feature_names)
```

```python
wine_data = wine.data
wine_target = wine.target
train_X,test_X,train_Y,test_Y = train_test_split(wine_data,wine_target,test_size = 0.2 )
print(len(train_X),len(train_Y))
print(len(test_X),len(test_Y))
```

```python
train_X = torch.from_numpy(train_X).float()
train_Y = torch.from_numpy(train_Y).long()

test_X = torch.from_numpy(test_X).float()
test_Y = torch.from_numpy(test_Y).long()
train_X = train_X[:len(train_X)]
test_X = train_X[:len(test_X)]

print(test_X.shape)
print(test_Y.shape)
print(train_X.shape)
print(train_Y.shape)
```

```python
train = TensorDataset(train_X, train_Y)

train_loader = DataLoader(train,batch_size=16,shuffle=True)
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.fc1 = nn.Linear(13,96)
        self.fc2 = nn.Linear(96,3)
    def forward(self, x):
        x= function.relu(self.fc1(x))
       
        x =self.fc2(x)
        return function.softmax(x)
model = Net()
```

```python
print(model)
```

```python

criterion =nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(),lr = 0.01)
for epoch in range(300):
    total_loss= 0
    for train_X,train_Y in train_loader:
        train_X,train_Y = Variable(train_X), Variable(train_Y)
        optimizer.zero_grad()
        output = model(train_X)
        loss = criterion(output,train_Y)
        loss.backward()
        optimizer.step()
        total_loss += loss.data[0]
    if(epoch+1) % 50 == 0:
         print(epoch+1, total_loss)
```

```python
test_X,test_Y = Variable(test_X),Variable(test_Y)
print(test_X)
result = torch.max(model(test_X).data,1)[1]
print(torch.max(model(test_X).data, 1))
#print(model(test_X).data)
accuracy = sum(test_Y.data.numpy()==result.numpy()) / len(test_Y.data.numpy())
print(result.numpy())
print(test_Y.data.numpy())
# ëª¨ë¸ì—ì„œ ë‚˜ì˜¨ê²ƒê³¼ resultê°€ ì¼ì¹˜í•˜ëŠ” ê°œìˆ˜ë¥¼ êµ¬í•œë‹¤.
print(accuracy)
```
